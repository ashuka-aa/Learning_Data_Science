#### Line Plot

```python
x = np.arange(1,11)
x
```

```python
y = 2 * x
y
```

```python
plt.plot(x,y)
plt.show()
```

#### Adding titles and labels

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-Label')
plt.ylabel('y-Label')
plt.show()
```

#### Changing line asethetic

```python
plt.plot(x,y,color = 'red',linestyle = ':',linewidth = 5)
plt.title('Line Plot')
plt.xlabel('x-Label')
plt.ylabel('y-Label')
plt.show()
```

```python
plt.plot(x,y,color = 'red',linestyle = '--',linewidth = 2)
plt.title('Line Plot')
plt.xlabel('x-Label')
plt.ylabel('y-Label')
plt.show()
```

```python
# linestyle -> dotted, dash, solid, dashdot
```

```python
plt.figure(figsize =(7,7)) #700x700 px resolution
plt.plot(x,y,color = 'red',linestyle = 'solid',linewidth = 2, marker = '>', markersize = 10)
plt.title('Line Plot')
plt.xlabel('x-Label',fontsize = 14, fontweight = 'bold', fontstyle = 'italic')
plt.ylabel('y-Label')
plt.show()
```

```python
plt.figure(figsize =(7,7)) #700x700 px resolution
plt.plot(x,y,color = 'red',linestyle = 'solid',linewidth = 2, marker = '>', markersize = 10, label ='Area Mapping')
plt.title('Line Plot')
plt.xlabel('x-Label',fontsize = 14, fontweight = 'bold', fontstyle = 'italic')
plt.ylabel('y-Label')
plt.legend(loc = 'upper center') #shows the label
plt.grid(True, linestyle = '--', alpha = 1)
plt.show()
```

```python
y1 = 3*x
y2 = np.exp(x)
```

```python
plt.figure(figsize =(7,7))
plt.plot(
    x,y,color = 'red',
    linestyle = 'solid',
    linewidth = 2, marker = '>', 
    markersize = 10, label ='Area Mapping'
)

plt.plot(
    x,y1,color = 'purple',
    linestyle = 'dotted',
    linewidth = 2, marker = 'd', 
    markersize = 12, label ='Area Mapping'
)

plt.plot(
    x,y2,color = 'green',
    linestyle = 'dotted',
    linewidth = 2, marker = 'p', 
    markersize = 12, label ='Area Mapping'
)

plt.title('Line Plot')
plt.xlabel('x-Label',fontsize = 14, fontweight = 'bold', fontstyle = 'italic')
plt.ylabel('y-Label')
plt.legend(loc = 'upper left', bbox_to_anchor = (1,1)) 
plt.grid(True, linestyle = '--', alpha = 1)
plt.show()


# bbox_to_anchor = (2,1) -> moves legend relative to the graph box (inside or outside)
# loc decides which point of the legend box is used
# bbox_to_anchor moves that point to a specific coordinate
```

### Figure and Axes Objects

```python
# fig -> the whole figure
# axes -> individual subplot boxes

fig,axes = plt.subplots(3,3, figsize= (7,7)) # makes a grid of subplots.

for i in range(3):
    for j in range(3):
        axes[i,j].plot(x,y)
        axes[i,j].set_title('Line Plots')
        axes[i,j].set_xlabel('x-axis')
        axes[i,j].set_ylabel('y-axis')

plt.tight_layout()
plt.show()
        
```

```python
fig,axes = plt.subplots(3,3, figsize= (7,7))
axes = axes.flatten()

for i in range(9):
        axes[i].plot(x,y)
        axes[i].set_title('Line Plots')
        axes[i].set_xlabel('x-axis')
        axes[i].set_ylabel('y-axis')

plt.tight_layout()
plt.show()
        
```
