### Qn: Create a 5x5 subplot grid and plot different plots only on the border and diagonal cells while hiding the rest.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Line Plot

def one(i,j):
    x = np.array([1,2,3,4,5])
    y = np.array([2,3,4,5,6])
    
    axes[i][j].plot(x,y, linestyle = 'dotted', color = 'red', marker = 'o', label = 'x and y')
    axes[i][j].set_title('Plotting X and Y')
    axes[i][j].set_xlabel('X-Axis')
    axes[i][j].set_ylabel('Y-Axis')
    axes[i][j].grid(True, linestyle = 'dashed', alpha = 0.7)
    axes[i][j].legend(loc = 'upper left')

# Bar Graph

def two(i,j):
    df = pd.DataFrame({
    'Name' : ['Ashuka','Thomas','Dipesh','Sudip'],
    'Marks': [45,55,62,29]
    })
    
    bars = axes[i][j].bar(
        df['Name'],
        df['Marks'], 
        linestyle = 'dotted', 
        color = ['pink','purple','yellow','green'], 
        edgecolor = 'black',
        linewidth = 1,
        label = 'Student_Details'
    )
    for bar in bars:
        height = bar.get_height()

        axes[i][j].text(
            bar.get_x() + bar.get_width()/2,
            height,
            f'{height}',
            ha = 'center',
            va = 'bottom',
            fontsize = 10,
            fontweight = 200
        )     
    axes[i][j].set_title('Student Details')
    axes[i][j].set_xlabel('Students')
    axes[i][j].set_ylabel('Marks')
    axes[i][j].tick_params(axis = 'x', rotation = 45)
    axes[i][j].grid(True, linestyle = 'dashed', alpha = 0.7)

# Scatter Plot

def three(i,j):
    x = np.array([10,20,30,40,50,60,70,80,90])
    a = np.array([8,1,7,2,0,3,7,3,2])

    axes[i][j].scatter(x, a, marker='*', c= 'purple', s = 100)
    axes[i][j].set_xlabel('Values of x')
    axes[i][j].set_ylabel('Values of a')
    axes[i][j].set_title('Scatter Plot')

# Horizontal Bar Graph

def four(i,j):
    df = pd.DataFrame({
    'Fruits' : ['Apple','Grapes','Banana','Kiwi'],
    'Quantity': [10,32,12,9]
    })

    bars = axes[i][j].barh(
        df['Fruits'], 
        df['Quantity'],
        color = ['red','lightgreen','yellow','brown'],
        edgecolor = 'black',
        linewidth = 1,
        label = 'Fruits'
    )

    for bar in bars:
        width = bar.get_width()

        axes[i][j].text(
            width,
            bar.get_y() + bar.get_height()/2,
            f'{width}',
            ha = 'left',
            va = 'center',
            fontweight = 'bold'
        )
    axes[i][j].set_title('Fruits Detail')
    axes[i][j].set_xlabel('Quantities')
    axes[i][j].set_ylabel('Fruits')
    axes[i][j].legend(loc = 'upper right')

# Box Plot

def five(i, j):
    first = [1,2,3,4,5,6,7,8,9,20]
    second = [1,2,3,4,5,6,7,-20]
    third = [2,3,6,4,8,25]

    sns.set_style('whitegrid')

    sns.boxplot(data=[first, second, third], palette='Set2', ax=axes[i][j])

    axes[i][j].set_title('Data')
    axes[i][j].set_xlabel('x-axis')
    axes[i][j].set_ylabel('y-axis')

# Violin Plot

def six(i,j):
    a = [2,3,4,5,6,90]
    b = [23,5,6,7,8,9]
    c = [12,45,23,77,99,-14]
    
    axes[i][j].violinplot([a,b,c], showmedians = True)

# Pie-Chart

def seven(i,j):
    df = pd.DataFrame({
        'Name' : ['Ram', 'Hari', 'Sita', 'Gita','Rita'],
        'Marks' : [28,90,12,78,98]
    })

    axes[i][j].pie(
        df['Marks'], 
        labels = df['Name'], 
        autopct = '%0.1f%%',
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 1},
        textprops = {'fontsize' : 8, 'fontweight' : 300 },
        colors = ['blue','lightgreen','yellow','brown','pink']
    )
    axes[i][j].set_title('Student Details', fontweight = 400)
    
# Exploded Pie-Chart

def eight(i,j):
    df = pd.DataFrame({
    'Activity': ['Sleep', 'Work', 'Exercise', 'Leisure', 'Study'],
    'Hours': [8, 9, 1, 3, 3]
    })
    explodes = [0.1]* len(df['Activity'])

    axes[i][j].pie(
        df['Hours'], 
        labels = df['Activity'] , 
        autopct = '%0.1f%%',
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 1},
        textprops = {'fontsize' : 8, 'fontweight' : 300 },
        colors = ['pink', 'lightblue', 'lightgreen', 'purple', 'brown'],
        explode = explodes,
        shadow = True
    )
    axes[i][j].set_title('Daily Activities Distribution', fontweight = 400)
    
# Donut Chart

def nine(i,j):
    df = pd.DataFrame({
    'Activity': ['Sleep', 'Work', 'Exercise', 'Leisure', 'Study'],
    'Hours': [8, 9, 1, 3, 3]
    })
    explodes = [0.1]* len(df['Activity'])

    axes[i][j].pie(
        df['Hours'], 
        labels = df['Activity'] , 
        autopct = '%0.1f%%',
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 1},
        textprops = {'fontsize' : 8, 'fontweight' : 300 },
        colors = ['red', 'blue', 'green', 'orange', 'purple'],
        startangle = 90,
        radius = 1
    )
    axes[i][j].set_title('Daily Activities Distribution', fontweight = 400)
    axes[i][j].pie([6],colors = 'white', radius = 0.55)
    axes[i][j].text(0,0,'Time Distribution', ha = 'center', va = 'center', fontweight = 'bold')
    
# Histogram

def ten(i,j):
    df = pd.DataFrame({
    'Age': [22, 25, 19, 24, 30, 28, 35, 21, 26, 27,
            29, 23, 24, 22, 21, 31, 32, 20, 23, 25,
            26, 28, 27, 24, 22, 23, 29, 30, 31, 22]
    })

    sns.histplot(np.log1p(df['Age']),ax=axes[i][j], kde = True)

    axes[i][j].set_title('Age of Students')
    axes[i][j].set_xlabel('Age')
    axes[i][j].set_ylabel('Frequency')
    axes[i][j].grid(True, linestyle = '--', alpha = 0.7)

# Polar Plot

def eleven(i,j):
    axes[i][j].remove()

    polar_axes = fig.add_subplot(5,5, i * 5 + j + 1, projection = 'polar')

    r = np.linspace(0,10,100)
    theta = 2 * np.pi * r

    polar_axes.plot(theta, r)
    polar_axes.set_title('Polar Plot')

# HexBin Plot

def twelve(i,j):
    x = np.random.randn(1000)
    y = np.random.randn(1000)
    
    hb = axes[i][j].hexbin(x,y, gridsize = 30, cmap= 'Blues')
    axes[i][j].set_title('HexBin Plot')
    axes[i][j].set_xlabel('x-axis')
    axes[i][j].set_ylabel('y-axis')
    fig.colorbar(hb, ax = axes[i][j],label = 'Count')
    
# Contour Graph

def thirteen(i,j):
    x = np.linspace(-3.0,3.0,100)
    y = np.linspace(-3.0,3.0,100)

    X,Y = np.meshgrid(x,y)

    Z = X**2 + Y**2
    
    cg = axes[i][j].contour(X,Y,Z)

    axes[i][j].set_title('Contour Mapping')
    axes[i][j].set_xlabel('x-axis')
    axes[i][j].set_ylabel('y-axis')

    fig.colorbar(cg, ax = axes[i][j], label = 'Z')

# 3D Scatter Plot

def fourteen(i,j):
    axes[i][j].remove()

    x = np.random.randn(100)
    y = np.random.randn(100)
    z = np.random.randn(100)

    scatter_3D = fig.add_subplot(5,5, i*5 + j + 1, projection = '3d')

    scatter_3D.scatter(x,y,z)
    scatter_3D.set_title('3D Scatter Plot')
    scatter_3D.set_xlabel('x-axis')
    scatter_3D.set_ylabel('y-axis')
    scatter_3D.set_zlabel('z-axis')
    
```

```python
fig, axes = plt.subplots(5,5,figsize = (20,20))

n = 5
list1 = [one, two, three, four, five, six, seven, eight, nine, ten, eleven, twelve, thirteen, fourteen]
k = 0

for i in range(n):
    for j in range(n):
        if i == 0 or i == 4:
            list1[k](i,j)
            k = (k + 1) % len(list1)
            
        elif j ==0 or j == 4:
            list1[k](i,j)
            k = (k + 1) % len(list1)
            
        elif i == j or i + j + 1 == 5:
            list1[k](i,j)
            k = (k + 1) % len(list1)
            
        else:
            axes[i,j].set_visible(False)

plt.tight_layout()
plt.savefig('Plots.png')
plt.show()
    
```
