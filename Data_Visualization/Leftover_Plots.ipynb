### Polar Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
r = np.linspace(0,10,100)
theta = 2 * np.pi * r

fig = plt.figure(figsize = (7,7))

ax = fig.add_subplot(111,projection = 'polar')

plt.plot(    # same as ax.plot cause we only have one axes to plot on so matplotlib automatically picks it 
    theta,
    r,
    linewidth = 2,
    marker = 'o',
    markersize = 3,
    color = 'purple'
)
plt.title('Polar Plot using plt.plot')
plt.show()
```

```python
directions = np.radians([0, 45, 90, 135, 180, 225, 270, 315])  
speed = [3, 7, 5, 2, 8, 4, 6, 1]  

ax = plt.subplot(111, projection = 'polar')

ax.bar(
    directions,
    speed,
    color = 'lightgreen',
    width = 0.5
)
ax.set_xticks(directions)

ax.set_xticklabels(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW'])

ax.set_title('Polar Plot using plt.bar')

plt.show()



```

```python
theta = np.linspace(0, 2 * np.pi, 100)
r     = 1 + 0.5 * np.sin(3 * theta)  

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.fill(theta,r, color = 'lightblue', alpha= 0.4)
plt.show()
```

### Hexbin Plot

```python
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hexbin(x,y,cmap = 'Greens', gridsize = 30)
plt.colorbar(label = 'count')

plt.title('HexBin Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```

### Contour Mapping

```python
x = np.linspace(-3.0,3.0,100)
y = np.linspace(-3.0,3.0,100)

X,Y = np.meshgrid(x,y)

Z = np.sin(X**2 + Y**2)
```

```python
plt.contour(X, Y, Z, levels=20, cmap='plasma')
plt.colorbar(label= 'Z-Values')

plt.title('Cotour Mapping')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```
