```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Pie-Chart

```python
fruits = {
    'Apple': 20,
    'Watermelon': 30,
    'Banana': 12,
    'Kiwi': 16,
    'Grapes': 200,
    'Mango': 5,
}
```

```python
key = fruits.keys()
value = fruits.values()
```

```python
print(f'Fruits = {list(key)}')
print(f'Quantity = {list(value)}')
```

```python
plt.figure(figsize = (10,10))

plt.pie(value, 
        labels = key,
        autopct = '%0.01f%%',
        colors = ['red', 'blue', 'pink', 'orange','brown','grey'],
        startangle = 90
       )
plt.title('Fruits Distribution', fontweight = 'bold', fontsize = 16)
plt.legend(labels = key, title = 'Fruits', loc = 'upper right')
plt.show()





```

```python
# using wedges, text, autotext

# wedges -> the slices
# text -> list of label 
# autotext -> list of percentage

plt.figure(figsize = (10,10))

wedges, text, autotext = plt.pie(
    value,
    labels = key,
    autopct = '%0.01f%%',
    startangle = 90,
    colors = sns.color_palette('Set2', len(fruits)),
    wedgeprops = {'edgecolor' : 'black', 'linewidth' : 1},
    textprops = {'fontsize' : 8, 'weight': 'bold'}
)
for i in autotext:
    i.set_color('white')

plt.title('Fruits Distribution', fontweight = 'bold')
plt.legend(wedges, key, loc = 'upper right', bbox_to_anchor = (1,1), title = 'Fruits')
plt.show()
```

### Exploded Pie-Chart

```python
plt.figure(figsize = (10,10))

explodes = [0.1] * len(fruits)

wedges, text, autotext = plt.pie(
    value,
    labels = key,
    startangle = 90,
    shadow = True,
    autopct = '%0.01f%%',
    explode = explodes,
    wedgeprops = {'edgecolor':'black'},
    textprops = {'weight':400, 'fontsize': 10},
    colors = sns.color_palette('Set1',len(fruits)),
    radius = 1
)
plt.legend(wedges, key, loc = 'upper right', title = 'Fruits')
plt.title('Exploded Pie-Chart', fontweight = 'bold')

for i in autotext:
    i.set_color('white')

plt.show()



```

### Donut Chart

```python
plt.figure(figsize = (10,10))

explodes = [0.1] * len(fruits)

wedges, text, autotext = plt.pie(
    value,
    labels = key,
    startangle = 90,
    autopct = '%0.01f%%',
    wedgeprops = {'edgecolor':'black'},
    textprops = {'weight':400, 'fontsize': 10},
    colors = sns.color_palette('Set3',len(fruits)),
    radius = 1
)
plt.pie([len(fruits)], colors = 'white', radius = 0.5)
plt.legend(wedges, key, loc = 'upper right', title = 'Fruits')
plt.title('Donut Chart', fontweight = 'bold')


plt.show()



```

### Scatter Plot

```python
x = np.array([10,20,30,40,50,60,70,80,90])
a = np.array([8,1,7,2,0,3,7,3,2])
```

```python
plt.scatter(x,a, marker = 'p', color = 'red', s = 100, edgecolor = 'black',linewidth = 0.8, alpha = 0.7)
plt.title('Scatter Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
z = np.array([10,20,30,40,50,60,70,80,90])
x = np.array([8,1,7,2,0,3,7,3,2])
y = np.array([7,2,5,6,9,1,4,5,3])
```

```python
plt.scatter(z,x, marker = 'o', s = 200, edgecolor = 'black', color = 'orange', linewidth = 0.8)
plt.scatter(z,y, marker = '*', s = 100, edgecolor = 'black', linewidth = 0.8, color = 'blue')
plt.title('Scatter Plots')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

### Adding Two Plots

```python
plt.figure(figsize = (9,9))

plt.plot(z,x, linestyle = 'dashed', linewidth = 1.2, color = 'red')
plt.scatter(z,x, marker = 'd', color = 'blue', edgecolor = 'black', linewidth = 0.7)

plt.title('Line Plot along with Scatter Plot', fontweight = 'bold')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```
