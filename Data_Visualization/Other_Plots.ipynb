```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Box Plot

```python
one = [1,2,3,4,5,6,7,8,9,20]
two = [1,2,3,4,5,6,7,-20]
three = [2,3,6,4,8,25]
```

```python
plt.figure(figsize = (10,10))

sns.set_style('darkgrid')

sns.boxplot(data = [one,two,three],
            palette = 'Set1',
            linewidth = 1.5,
            width = 0.4,
            flierprops = {'marker':'*', 'markersize':10, 'markerfacecolor':'red' }
           )
plt.title('Box-Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# using the dataframe in its wide format

plt.figure(figsize = (10,10))

sns.set_style('darkgrid')

sns.boxplot(
    data = df,
    palette ='Set2',
    width = 0.5,
    linewidth = 1.5,
    flierprops = {'marker' : 'o', 'markerfacecolor' : 'blue', 'markersize':10}
)
plt.xlabel('Groups')
plt.ylabel('Values')
plt.savefig('Box.png')
plt.show()
```

### Violin Plot

```python
plt.figure(figsize = (10,10))

sns.set_style('darkgrid')

plt.violinplot([one,two,three], showmedians = True)
plt.title('Violin Plot using Matplotlib')
plt.savefig('ViolinMat.png')
plt.show()
```

```python
plt.figure(figsize = (10,10))

sns.set_style('darkgrid')

sns.violinplot(
    data = df_long,
    x = 'Group',
    y = 'Values',
    hue = 'Group',
    width = 0.5,
    palette = 'Set2',
    linewidth = 1.5
)
plt.title('Violin Plot using Seaborn')
plt.savefig('Violin.png')
plt.show()
```

### Histogram

```python
values = np.random.normal(50,200,100)
```

```python
plt.figure(figsize = (10,10))

sns.histplot(
    data = values,
    kde = True,
    bins = 20,
    color = 'brown',
    edgecolor = 'black',
    linewidth = 0.5
)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Count')
plt.savefig('Hist.png')
plt.show()
```

### 3D Scatter Plot

```python
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)
```

```python
fig = plt.figure()

ax = fig.add_subplot(111, projection = '3d')

ax.scatter(x,y,z, color = 'purple')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')
plt.savefig('3D.png')
plt.show()
```
