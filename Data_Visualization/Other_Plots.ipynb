```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Box Plot

```python
one = [1,2,3,4,5,6,7,8,9,20]
two = [1,2,3,4,5,6,7,-20]
three = [2,3,6,4,8,25]
```

```python
plt.figure(figsize = (10,10))

sns.set_style('darkgrid')

sns.boxplot(data = [one,two,three],
            palette = 'Set1',
            linewidth = 1.5,
            width = 0.4,
            flierprops = {'marker':'*', 'markersize':10, 'markerfacecolor':'red' }
           )
plt.title('Box-Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```
