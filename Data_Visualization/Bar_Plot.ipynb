### Bar PLot

```python
import os
import math
import re
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
students= {
    'Ashuka Acharya' : 20,
    'Simran KC' : 78,
    'Sudip Karki': 68,
    'Aayushma Nepal':90
}
students
```

```python
type(students)
```

```python
names = list(students.keys())
marks = list(students.values())
```

```python
print(names,marks)
```

```python
bars = plt.bar(
    names, 
    marks,
    color = ['red','yellow','blue'],
    edgecolor = 'black',
    linewidth = 1
)

plt.title('Bar Plot', fontsize = 16, fontweight = 'bold')
plt.xlabel('Name of students')
plt.ylabel('Marks of students')
plt.xticks(rotation = 90)
plt.grid(axis = 'y', linestyle = '--', alpha = 1)

# To write the marks of each student on top of each bars

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'{height}',
        ha = 'center',
        va = 'bottom',
        fontsize = 10,
        fontweight = 100
    )
    
plt.show()
```

### Horizontal Bar Chart

```python
bars = plt.barh(
    names, 
    marks,
    color = ['red','yellow','blue'],
    edgecolor = 'black',
    linewidth = 1
)

plt.title('Horizontal Bar Plot', fontsize = 16, fontweight = 'bold')
plt.xlabel('Marks of students')
plt.ylabel('Name of students')

plt.grid(axis = 'y', linestyle = '--', alpha = 1)

# To write the marks of each student at the end of each bar

for bar in bars:
    width = bar.get_width()

    plt.text(
        width,
        bar.get_y() + bar.get_height()/2,
        f'{width}',
        ha = 'left',
        va = 'bottom',
        fontsize = 10,
        fontweight = 100
    )
    
plt.show()
```

#### Alternate way to add the text using plt.bar_label()

```python
bars = plt.bar(
    names, 
    marks,
    color = ['red','yellow','blue'],
    edgecolor = 'black',
    linewidth = 1
)

plt.title('Bar Plot', fontsize = 16, fontweight = 'bold')
plt.xlabel('Name of students')
plt.ylabel('Marks of students')
plt.xticks(rotation = 90)
plt.grid(axis = 'y', linestyle = '--', alpha = 1)

# To write the marks of each student on top of each bars

plt.bar_label(bars, label_type = 'edge', padding = 3, fontsize = 10, fontweight = 500, color = 'grey')
    
plt.show()
```

### Adding Error Bars

```python
bars = plt.bar(
    names, 
    marks,
    color = ['red','yellow','blue'],
    edgecolor = 'black',
    linewidth = 1,
    yerr=[2,3,1,2]
)

plt.title('Bar Plot', fontsize = 16, fontweight = 'bold')
plt.xlabel('Name of students')
plt.ylabel('Marks of students')
plt.xticks(rotation = 50)
plt.grid(axis = 'y', linestyle = '--', alpha = 1)
plt.ylim(0,100) # setting limits

# To write the marks of each student on top of each bars
plt.bar_label(bars, label_type = 'edge', padding = 3, fontsize = 10, fontweight = 500)
    
plt.show()
```

### Stacked Bar Chart

```python
name = ['Ram', 'Hari', 'Sita', 'Gita']
marks_math = np.array([10,24,67,89])
marks_science = np.array([90,78,56,34])
```

```python
bars_math = plt.bar(
    name, 
    marks_math, 
    label ='Maths', 
    color = 'grey'
    )

bars_science = plt.bar(
    name, 
    marks_science, 
    bottom = marks_math, 
    label = 'Science',
    color = 'brown'
)

for bars in bars_math:
    height = bars.get_height()

    plt.text(
        bars.get_x() + bars.get_width()/2,
        height,
        f'{height}',
        ha = 'center',
        va = 'bottom'
    )

    
for math, science, x in zip(marks_math, marks_science,bars_math):
    total = math + science
    
    plt.text(
        x.get_x() + x.get_width()/2,
        total,
        f'{total}',
        ha = 'center',
        va = 'bottom'
    )
    
plt.xlabel('Students')
plt.ylabel('Marks')

plt.legend()
plt.show()
```

#### Grouped Bar Chart

```python
x = np.arange(len(name))  # [0,1,2,3]

width = 0.4 # Width of each bar

plt.bar(x, marks_math, width=width, label='Math', color='grey')
plt.bar(x + width, marks_science, width=width, label='Science', color='lightgreen')

plt.xticks(x + width/2, name)

plt.xlabel("Students")
plt.ylabel("Marks")

plt.title("Grouped Bar Chart")
plt.legend(bbox_to_anchor= (1,1))
plt.show()
```
