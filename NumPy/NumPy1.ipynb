### Adding NumPy Array

```python
a = np.array([1,2,3])
b = np.array([13,14,15])

print (np.sum([a,b])) #adds all elements
print ('Axis = 0',np.sum([a,b] ,axis=0)) # axis=0 —> sums down each column
print ('Axis = 1',np.sum([a,b] ,axis=1)) # axis=1 —> sums across each row
print (np.subtract(a,b))
```

### Multipying NumPy Array

```python
a = np.array([1,2,3])
b = np.array([13,14,15])

print (np.prod([a,b])) #adds all elements
print ('Axis = 0',np.prod([a,b] ,axis=0)) # axis=0 —> sums down each column
print ('Axis = 1',np.prod([a,b] ,axis=1)) # axis=1 —> sums across each row
```

#### Qn: Create a NumPy array with data ranging from 10 to 1000, calculate the covariance using both np.cov() and from scratch, then compare the results.

```python
x = np.random.randint(10,1000,20)
y = np.random.randint(10,1000,20)

print(f'x = {x}') 
print(f'y = {y}') 
```

##### Manually from scratch

```python
x_total = 0
y_total = 0

n = len(x)
for i in range(n):
    x_total += x[i]
    y_total += y[i]

x_mean = x_total/len(x)
y_mean = y_total/len(y)

total = 0

for j in range(n):
    total += ((x[j] - x_mean) * (y[j] - y_mean))

m_cov = total / n

print(f'The covariance of x and y (calculated manually): {m_cov}')
    
```

##### Using NumPy (np.cov())

```python
# np.cov() gives covariance matrix from which we extract cov(x,y) using [0,1]
# [[cov(x,x),  cov(x,y)],     
#  [cov(y,x),  cov(y,y)]]

np_cov = np.cov(x,y)[0,1]

print(f'The covariance of x and y (calculated using NumPy): {np_cov}')
```

##### Difference between the two and the reason why?

```python
print(f'The difference between the covariances calculated manually and using NumPy: {abs(m_cov - np_cov)} ')
```

##### The reason why the two values are different is that np.cov() calculates sample covariance i.e. using n-1 but manually, we calculated population covariance i.e using N.

```python
# Population Covariance can also be calculated using NumPy

pop_cov = np.cov(x,y,ddof = 0)[0,1]

print(f'The population covariance of x and y calculated using NumPy : {pop_cov}')
```

```python
print(f'The diff between the covariances calculated manually and NumPy(population):{abs(m_cov-pop_cov)}')
```
