### cancel
It uses factoring internally as the first step. And then goes one step further by cancelling the common terms.  
It is faster than simplify as it is only used for a specific purpose i.e only cancels fractions  
But, simplify is used for general purposes and tries many strategies internally, which makes it slower.  
Also, cancel cannot be used with trigonometric expressions.

```python
exp = (x**2 - 1)/(x - 1)

sp.cancel(exp)
```

```python
expr = (x**2 + 2*x) / x

sp.cancel(expr)
```

```python
expr = (x + 1) / (x + 2)

sp.cancel(expr) # nothing to cancel
```

### factor

It breaks down an expression into its multiplied components. It is used to find the roots of expressions, simplify complex equations, and break down polynomials.

```python
exp1 = x**2 - x - 2

sp.factor(exp1)
```

```python
a, b = sp.symbols('a b')

exp2 = a**2 + 2*a*b + b**2

sp.factor(exp2)
```

### expand
It opens up and multiplies out any expression including brackets, powers, trig, logs. It is the foundation of algebraic manipulation and is heavily used in DS/AI.  
expand() and factor() are complete opposites of each other.   

```python
y = sp.Symbol('y')

exp1 = (x - y)**2

sp.expand(exp1)
```

```python
# trignometric expressions

e = sp.sin(x + y)

sp.expand(e, trig = True)
```

```python
# logarithmic expressions

e1 = sp.log(x * y)

sp.expand(e1, force = True)

# SymPy is cautious, won't expand logs until it is sure that the variables are positive. 
# So force = True, tells SymPy that the variables are positive and it is safe.
```

```python
# Alternative: Declare symbols as positive

a, b = sp.symbols('a b', positive = True)

e2= sp.log(a * b)

sp.expand(e2)
```

Piecewise means different formulas depending on a condition.   
General form:  
$$
Piecewise((expression1, condition1),
          (expression2, condition2))
$$
In our result, 
$$
                Piecewise((a**x/log(a), Ne(log(a), 0)), (x, True)) + C
$$
If log(a) != 0, then,  
$$
result = a**x/log(a) 
$$
(Ne means Not Equal)  
else, 
$$
result = x 
$$
True acts like the default case.

### lamdify
It converts a SymPy symbolic expression into a regular Python function that can be used for fast numerical computation.  
It is the bridge between symbolic math and numerical computation.  
We derive expressions symbolically in SymPy, then convert them to fast NumPy functions for actual computation, plotting, and use in DS/AI workflows.  
Syntax: sp.lambdify(variable, expression, library) -> default library is 'numpy'  

Libraries:
* Always default to 'numpy' in DS/AI —> it handles arrays, is fast, and works with plotting
* Use 'math' only for simple single value calculations
* Use 'scipy' only when you need special scientific functions not in numpy

```python
exp = 5*x + 2
exp(2) # as in exp, the variable x is just a symbol so we cannot pass any values, so we use lamdify to do so
```

```python
f = sp.lambdify(x, exp, 'math')
f(5)
```

```python
# arrays acan also be passed

f = sp.lambdify(x, exp, 'math')

f(np.array([1,2,3,4,5]))
```

```python
# we can also substitute a single value using .subs() (not arrays though)

exp = 6 * x
exp.subs(x,2)
```

```python
# for mutiple variables

exp1 = x + y
exp1.subs([(x,2),(y,7)])
```

### quad
It numerically integrates a function over a given interval, finds the area under a curve between two points.  
It returns two values, the result and the estimated error.  
integrate.quad() needs a real callable function to pass values into.(use lambdify)

```python
exp = x ** 3 + 3

f = sp.lambdify(x, exp)

result, error = integrate.quad(f, 0,1)

print(f'The result = {result} and error = {error}')
```

### solve
It finds the values of a variable that makes an expression equal to zero, or solves an equation.


```python
exp = (x + 2) # solves x + 2 = 0

sp.solve(exp, x)
```

```python
exp1 = x**2 + 5*x + 6

sp.solve(exp1, x)
```

```python
eq = sp.Eq(x - 3, 5) # it is equal to x - 3 = 5

sp.solve(eq, x)
```

```python
eq1 = sp.Eq(x**2, 4)

sp.solve(eq1, x)
```

```python
e1 = x**3 - 6*x**2 + 11*x - 6

sp.solve(e1, x)
```

```python
eq1 = sp.Eq(x + y, 5) 
eq2 = sp.Eq(x - y, 1) 

sp.solve([eq1, eq2], [x,y])
```

```python
sp.solve(x**2 + 1, x)  # no real solutions
```

```python
a, b, c = sp.symbols('a b c')

sp.solve(a*x**2 + b*x + c, x) # give us the quadratic formula
```

### limit
It finds the value an expression approaches as a variable gets closer and closer to a specific value, without necessarily reaching it.

```python
sp.limit(x + 2, x, 2) # as x approaches 2
```

```python
exp = sp.sin(x)/x

sp.limit(exp, x, 0)
```

```python
sp.limit(1/x, x, sp.oo) # as x approaches infinity
```

```python
sp.limit(1/x, x, -sp.oo) # as x approaches - infinity
```

```python
sp.limit(x + 2, x, sp.oo)
```

```python
# From the right (+)

sp.limit(1/x, x, 0, '+') 
```

```python
# From the left (-)

sp.limit(1/x, x, 0, '-') 
```
