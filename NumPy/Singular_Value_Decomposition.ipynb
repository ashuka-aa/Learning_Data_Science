### Singular Value Decomposition

### Calculated Manually

```python
import numpy as np

A = np.array([[3,1],[1,3]])

AT = A.T

X = AT @ A

print(f'A\u1D40 :\n{AT}')
print(f'A\u1D40.A :\n{X}')

evalue, evector = np.linalg.eig(X)

i = np.argsort(evalue)[::-1]  #sorting in descending order
evalue = evalue[i]
evector = evector[:,i]

print('Eigen Values of A\u1D40.A:',evalue)

print('Eigen Vector of A\u1D40.A:',evector,sep='\n')

singular_value = np.sqrt(evalue)

S = np.diag(singular_value)

print(f'Singular Vector (S) :\n{S}')

V = evector
VT = V.T

print(f'Right Singular Vector (V\u1D40) :\n{VT}')

m,n = A.shape
r = np.linalg.matrix_rank(A)

U = np.zeros((m,r),dtype = float)

for i in range(r):
    vi = evector[:,i]

    U[:,i] = (1/singular_value[i]) * (A @ vi)
 
print(f'Left Singular Vector (U) :\n{U}')

print(f'Reconstruction of A: \n{U @ S @ VT}')
```

### Calculated Using NumPy

```python
A = np.array([[3,1],[1,3]])

U,S_Values,V = np.linalg.svd(A)

S = np.diag(S_Values)

print(f'Left Singular Vector (U) :\n{U}')
print(f'Singular Vector (S) :\n{S}')
print(f'Right Singular Vector (V\u1D40) :\n{V}')

print(f'Reconstruction of A: \n{U @ S @ V}')
```
