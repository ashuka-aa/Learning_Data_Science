### Series Object in One Dimensional Array

```python
import pandas as pd

#using a list

s1 = pd.Series([1,2,3,4,5]) #list of values with labels attached to it

s1
```

```python
# custom index

s2 = pd.Series([1,2,3,4,5], index = ['a','b','c','d','e'])

s2
```

```python
# using a dictionary

s3 = pd.Series({
    'Name':'Ashuka',
    'Roll No':8,
    'Address': 'Naya Thimi',
    'Gender': 'Female'
})

s3
```

```python
# changing the position of an index

s4 = pd.Series({
    'a':1,
    'b':2,
    'c':3
},index = ['b','d','a','c'])

s4
```

#### Accessing Values

```python
s5 = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])

s5
```

```python
print(s5[0])

print()
print(s5[2:5])

# print(s5[-1]) -> Error

print()
print(s5['a'])
```

```python
# using .iloc and .loc

#iloc -> access by number position

print(s5.iloc[0])
print()
print(s5.iloc[0:2])
print()
print(s5.iloc[-1])
```

```python
#loc -> access by label/index name

print(s5.loc['a'])
print()
print(s5.loc['a':'d'])
print()
print(s5.loc['d':'a':-1])
```

#### Useful Operations

```python
a1 = pd.Series([1,2,3,4,50])
a2 = pd.Series([11,12,31,41,5])

a1 + a2
```

```python
a1 * a2
```

```python
a1 / a2
```

```python
a1 % a2
```
