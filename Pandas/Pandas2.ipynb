##### Qn: Increase salary by 15% for male with cgpa >= 3.5 into a new salary

```python
# checking the employeees who are male with cgpa >= 3.5

df[['Name','CGPA','Salary']][(df['Gender'] == 'Male') & (df['CGPA'] >= 3.5)]
```

```python
# Method - 1

def increase_sal(x):
    if (x['Gender'] == 'Male') & (x['CGPA'] >= 3.5):
        return x['Salary'] * 1.15
    else:
         return x['Salary']

df['Updated_Salary'] = df.apply(increase_sal,axis = 1)
```

```python
df
```

```python
df.drop(columns = 'Updated_Salary', inplace = True)
```

```python
df.head()
```

```python
# Method - 2

# .loc includes the last label -> inclusive
# .iloc excludes the last position -> exclusive

i = (df['Gender'] == 'Male') & (df['CGPA'] >= 3.5)

df.loc[i,'Salary'] *= 1.15

df
```

#### .iloc
.iloc works like a matrix, access any cell using df.iloc[row, column] where both row and column are position numbers starting from 0. Just like matrix coordinates, df.iloc[1, 3] means row 1, column 3. It only accepts numbers, never names or labels

#### .loc 
.loc works like a matrix/grid, access any cell using df.loc[row_label, column_name] where row is the index label and column is the column name. Just like df.loc[1, 'Salary'] means row with label 1, column named Salary. The only difference from .iloc is that it uses labels/names instead of position numbers


```python
# Method - 3 -> using lambda function

df['Salary'] = df.apply(lambda x: x['Salary'] * 1.15 if ((x['Gender'] == 'Male') & (x['CGPA'] >= 3.5)) else x['Salary'], axis = 1)
```

```python
df
```

```python
# Method - 4 -> using assign

df = df.assign(Salary = lambda x : np.where
         ( (x['Gender'] == 'Male') & (x['CGPA'] >= 3.5),  # condition
            x['Salary'] * 1.15, # if True
            x['Salary']  # if False    
         ))
```

```python
# np.where works like if/else
# np.where(condition, value_if_true, value_if_false)
```

apply() -> works on both a Series and DataFrame. When used on a Series like df['Salary'].apply(func) it passes one value at a time into the function. When used on a DataFrame like df.apply(func, axis=1) it passes one entire row at a time. Best used for complex logic that needs multiple columns. axis=0 sends each column into the function, axis=1 sends each row into the function.

assign() -> works only on a DataFrame. The entire DataFrame is passed as x into the lambda like df.assign(Salary = lambda x: x['Salary'] * 1.15). It is faster than apply because it works on the entire DataFrame at once instead of row by row. Best used for adding or updating columns. It always creates a new column and if you want to update an existig column just use the same column name. Also, it does NOT save to original DataFrame.

map() -> works only on a Series. It transforms each value one by one like df['Salary'].map(lambda x: x * 1.15). Its most powerful feature is dictionary mapping where you can replace values like df['Gender'].map({'Male': 1, 'Female': 0}). Best used for replacing or converting values. It does NOT save changes to the original DataFrame automatically

```python
# map

df['Gender'] = df['Gender'].map({
    'Female':1,
    'Male':0
})

df.head()
```
