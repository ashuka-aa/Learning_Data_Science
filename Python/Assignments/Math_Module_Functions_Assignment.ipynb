#### Qn:1
$$
f(x) = {\pi}x^3 + log(x) + \sqrt{x^3} + tan(x) + e^x
$$

```python
import math

x = 3 #given

e1 = math.pi * x**3
e2 = math.log(x)
e3 = math.sqrt(x**3)
e4 = math.tan(x)
e5 = math.exp(x)

fx = e1 + e2 + e3 + e4 + e5

print(f'The value of f(x) = {fx:.4f}')
```

#### Qn:2
$$
g(x) = \frac{e^{2x} + cosh(x)}{sinh(x) + log_{10}(x)} + \sqrt[3]{x^5} - tan^{−1}(x)
$$

```python
x = 2 #given

gx = (
    (math.exp(2*x) + math.cosh(x))/(math.sinh(x) + math.log10(x))
    + (x**5)**(1/3)
    - math.atan(x)
)

print(f'The value of g(x) = {gx:.4f}')
```

#### Qn:4
