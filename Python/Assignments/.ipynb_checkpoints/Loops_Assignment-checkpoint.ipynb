##### Qn:2 Implement factorial calculation for numbers up to 20 using only while loops. Display each multiplication step. Verify your result by calculating the sum of divisors of the factorial.

```python
n = 5 # n = 20 takes a lot of time to execute 

while n > 0:
    i = n
    fact = 1
    
    print('-'*30)
    print(f'Factorial of {n}'.center(30))
    print('-'*30)

    st = str(i)
    
    while i > 0:       
        fact *= i
        i -= 1
        if i != 0:
            st+= ' * ' + str(i)
            print(f'{st} = {fact}')
        else:
            # Verification
            j = fact
            s = 0
            print(f'The divisors of {fact}:')
            while j > 0:
                if fact % j == 0:
                    print(j)
                    s += j
                j -= 1
            print(f'The sum of divisor of {fact} = {s}')
            
    
    n-=1  
```

##### Qn: 3 Calculate the GCD of three numbers using only nested while loops and the Euclidean algorithm. Extend this to find the LCM of the same three numbers without using any multiplication (use only addition).

```python
n1 = int(input('Enter first number:'))
n2 = int(input('Enter second number:'))
n3 = int(input('Enter third number:'))

a = max(n1,n2)
b = min(n1,n2)

while b != 0:
    t = a % b
    a = b
    b = t

temp = a
a = max(temp,n3)
b = min(temp,n3)

while b != 0:
    t = a % b
    a = b
    b = t

print(f'The GCD of {n1},{n2} and {n3} = {a}')

m = max(n1,n2)
lcm = m

while lcm % n1 != 0 or lcm % n2 != 0:
    lcm += m

temp = lcm

while lcm % n3 != 0:
    lcm += temp

print(f'The LCM of {n1},{n2} and {n3} = {lcm}')
```

##### Qn:4 Write a program to determine if a number is prime. If prime, find the next prime number. If composite, find all its prime factors using trial division with optimization (only check up to sqrt(n)).

```python
n = int(input('Enter a number'))

if n <= 1:
    print(f'{n} is not a prime number')
elif n == 2 or n == 3:
    print(f'{n} is a prime number.')
else:
    prime = True
    for i in range(2,int(n ** (1/2))+1):
        if n % i == 0:
            prime = False
            break

if prime:
    print(f'{n} is a prime number.')

    next = n + 1
    
    while True:
        nprime = True

        for j in range(2,int(next ** (1/2)) + 1): 
            if next % j == 0:
                nprime = False
                next += 1
                break

        if nprime:
            print(f'The next prime number after {n} is {next}.')
            break
                
else:
    print(f'{n} is a composite number.')
    print(f'Prime factors of {n} are:')

    temp = n
    i = 2
    while temp > 1:
        while temp % i == 0:
            print(i)
            temp = temp // i
        i += 1
    
```

##### Qn: 8 Implement a program that checks if a given number is a perfect square without using square root function. Use binary search approach with while loop.


```python
num = int(input('Enter a number:'))
n = num
r1 = 0
r2 = n

while r1 <= r2:
    mid = (r1 + r2) // 2
    if mid ** 2 == num:
        break
        
    elif mid ** 2 < num:
        r1 = mid + 1

    elif mid ** 2 > num:
        r2 = mid - 1

if r1 > r2:
    print(f'{num} is not a perfect square')
else:
    print(f'{num} is a perfect square')
```

##### Qn:9 Find all Armstrong numbers (Narcissistic numbers) between 1 and 1000000. For each number, check if the sum of each digit raised to the power of the number of digits equals the number itself.

```python
print ('-'*20)
print('Armstrong numbers'.center(20)) # to ensure the text is in the center of the decoration
print ('-'*20)

for i in range(1,1000000+1):
    p = len(str(i)) # to determine the length of the number
    n = i
    result = 0
    while n != 0:
        r = n % 10
        result += (r ** p) # sum of digits raised to the power of the length of number
        n //= 10
    
    if result == i: 
        print(f'{i}') 
    
```

##### Qn:10 Determine if a number is: Perfect (sum of divisors = number), Abundant (sum > number), or Deficient (sum < number). Find all perfect numbers less than 10000.

```python
num = int(input('Enter a number:'))
dsum = 0

for i in range(1,num):
    if num % i == 0: 
        dsum += i   # adding all the proper factors of the number

if dsum == num:
    print(f'{num} is a perfect number')
elif dsum > num:
    print(f'{num} is a abundant number')
else:
    print(f'{num} is a deficient number')

print ('-' * 40)
print('Perfect numbers less than 10000'.center(40)) # to ensure the text is in the center of the decoration
print ('-' * 40)

ulimit = 10000
for i in range(2,ulimit):
    dsum = 1     # 1 is a proper divisor for all num > 1
    for j in range(2,int(i ** (1/2))+1):
        if i % j == 0:
            pair = i // j
            if pair == j:
                dsum += j
            else:
                dsum += (j + pair)
            
    if dsum == i:
        print(i)
```

##### Qn 11: Implement the Sieve of Eratosthenes without using arrays – use nested loops to mark and identify primes up to n = 10000. Count twin primes (primes differing by 2).

```python
n = 10000
last_prime = 0
twin_prime = 0

for i in range(2,n+1):
    prime = True
    
    for j in range(2,int(i**(1/2))+1):
        if i % j == 0:
            prime = False
            break
    
    if prime:
        print(i)
        
        if last_prime != 0 and i - last_prime == 2:
            print(f'Twin Prime: {i},{last_prime}')
            twin_prime += 1
    
        last_prime = i
        
print(f'The no of twin prime = {twin_prime}')
```

##### Qn: 12 Calculate Euler’s totient function \\(\varphi(n)\\) by counting numbers coprime to n. Then verify Euler’s theorem: \\(a^{\varphi(n)} \equiv 1 \pmod{n}\\) for coprime a and n.

```python
n = int(input('Enter the number:'))
i = 1
coprime_count = 0
j = 2
prime = True

while j * j <= n:
    
    if n % j == 0:
        prime = False
        break

    j += 1
    
if not prime:
    while i < n:
        a = n
        b = i
        
        while b != 0:
            a, b = b, a % b
            
        if a == 1:
            coprime_count += 1
    
        i += 1
else:
    coprime_count = n - 1

print(f"The Euler's totient function of {n} is {coprime_count}")

i = 1

while i <= n:
        a = n
        b = i
        
        while b != 0:
            a, b = b, a % b
 
        if a == 1: 

            if (i**coprime_count) % n == 1:
                print(f"Co-prime ({i},{n}) is verified using Euler's Theorem")

        i += 1
```

##### Qn 13: Find all Pythagorean triplets (a, b, c) where \\(a^2 + b^2 = c^2\\) and a + b + c = 1000. Use primitive triplet generation formulas with loops.

```python
max = 500
a,b,c = 0,0,0

for m in range(1,max + 1):
    if 500 % m != 0:
        continue
        
    n = (500 // m) - m
    
    if n <= 0:
        continue
        
    elif m < n:
        continue
        
    elif (m % 2 == 0 and n % 2 ==0) or (m % 2 != 0 and n % 2 !=0):
        continue
    
    a = m**2 - n**2
    b = 2*m*n
    c = m**2 + n**2
    
    break

if a**2 + b**2 == c**2 and a + b + c == 1000:
    print(f'Pythagorean Triplets whose sum is 1000 :{a}, {b} and {c}')
    
else:
    print('Pythagorean Triplets not found within the range!!!')
```

##### Qn 14: Implement the Chinese Remainder Theorem for two congruences using only loops and conditionals. Solve: \\(x \equiv a_1 \pmod{m_1}\\) and \\(x \equiv a_2 \pmod{m_2}\\).

##### Qn 18: Implement the Pollard’s rho algorithm for integer factorization using only loops. Find the smallest non-trivial factor of a composite number.

```python
import math

while True:
    n = int(input('Enter a number:'))
    
    prime = True
    
    for i in range(2,int(n**(1/2)) + 1):  #checking if the input is prime or not
        if n % i == 0:
            prime = False
            break

    if not prime:
        break

    elif n == 1: #since n is neither prime nor composite
        print('The value of n cannot be 1. Enter again!!!')

    else:
        print('Your input is prime. Enter again!!!')

x0 = int(input('Enter an initial value:'))  #seed

t = h = x0
store = [] #to store the non-trivial factors
temp = n

while n > 1:
        x0 = (x0**2 + 1) % n #xi = g(xi -1)
    
        t = (t**2 + 1) % n   #turtle
        h = (((h**2 + 1) % n)**2 + 1) % n   #hare
    
        d = math.gcd(abs(t-h),n)
       
        if d == 1:
            continue
            
        elif d == n:
            prime = True  #if d is prime then we append it to the list and break the loop
            
            for k in range(2,int(d**(1/2)) + 1):
                if d % k == 0:
                    prime = False
                    break

            if prime:
                store.append(d)
                break

            else:
                print("With this initial value and function, Pollard's Rho ALgorithm does not work.Incrementing seed:")
                x0 += 1.
                t = h = x0
                n = temp
                continue

        else:
            f = d
            prime = True
            
            for i in range(2,int(d**(1/2)) + 1):
                if d % i == 0:
                    prime = False
                    break

            if prime:    #if d is prime we can divide n with it and countiue
                while n % d == 0:
                    n //= d
                    store.append(d)

            else:    #if d is composite then we find the prime factors of it and then append to the list
                j = 2
                n //= d
                while f > 1:
                    while f % j == 0:
                        f //= j
                        store.append(j)
                    j += 1

print(f'The non-trivial factors of {temp} are:')

for x in sorted(store):
    print(x)
    
print(f'The smallest non-trivial factor: {min(store)}')
            
```

##### Qn 23: Create a spiral matrix of size n x n filled with numbers 1 to \\(n^2\\) in spiral order using only loops and directional conditionals.

```python
n = int (input('Enter a number:'))

l = [[None for i in range(n)] for j in range(n)]

top = 0
left = 0
right = n-1
bottom = n-1
r = 1

while top <= bottom and left <= right:
    for i in range(left,right + 1):
        l[top][i] = r
        r += 1
    top += 1
    
    for i in range(top,bottom + 1):
        l[i][right] = r
        r += 1
    right -= 1
    
    for i in range(right, left - 1,-1):
        l[bottom][i] = r
        r += 1
    bottom -= 1
    
    for i in range(bottom, top - 1, -1):
        l[i][left] = r
        r += 1
    left += 1

print(l)
```

##### Qn 34: Approximate \\(\pi\\) using Monte Carlo method simulation: generate random points in a unit square, count those inside a quarter circle, estimate \\(\pi \approx 4 \frac{\text{inside}}{\text{total}}\\). Use LCG for random number generation.

```python
h = int(input('Enter the x-coordinate of the center of the circle:'))
k = int(input('Enter the y-coordinate of the center of the circle:'))
r = int(input('Enter the radius of the circle:'))
print()

a = 4887
c = 20
x0 = 22
y0 =  29
m = 1007

total_points = 0
points_inside = 0
max = 100000

while total_points < max:
    x0 = (a * x0 + c) % m
    xresult = (h - r) + (2 * r) * (x0 / (m - 1))

    y0 = (a * y0 + c) % m
    yresult = (k - r) + (2 * r) * (y0 / (m - 1))

    if (xresult - h) ** 2 + (yresult - k) ** 2 <= r ** 2:
        points_inside += 1

    total_points += 1

pi = 4 * (points_inside/total_points)

print(f'The estimation of pi using Monte Carlo Simulation is {pi}')
    
```

##### Qn 37: Find the root of \\(f(x) = x^3 - 2x - 5 = 0\\) using bisection method. Iterate until \\(|f(x)| < 10^{-5}\\). Display all intermediate approximations.

##### Qn 45: Implement the secant method to solve \\(\tan(x) = x\\). Compare iterations needed vs bisection method.

```python
import math

tol = 10**-5
max_iter = 1000

x0 = 4.3
x1 = 4.7
s_iter = 0
x_next = 0
fx0 = math.tan(x0) - x0
fx1 = math.tan(x1) - x1

while s_iter < max_iter and abs(x1 - x0) > tol and abs(fx1) > tol:
    
    if (fx1 - fx0) == 0:
        break

    x_next = x1 - (fx1 * (x1 - x0) / (fx1 - fx0))
    s_iter += 1

    fx_next = math.tan(x_next) - x_next
    
    x0 = x1
    fx0 = fx1
    
    x1 = x_next
    fx1 = fx_next

a = 4.3
b = 4.7
b_iter = 0
c = (a + b)/2

fa = math.tan(a) - a
fb = math.tan(b) - b
fc = math.tan(c) - c

if fa * fb < 0:
    while b_iter < max_iter and abs(fc) > tol:

        if fc == 0:
            break
        
        elif fa * fc < 0:
            b = c
            fb = fc
            
        else:
            a = c
            fa = fc
            
        c = (a + b)/2
        fc = math.tan(c) - c
        
        b_iter += 1

else:
    print(f'[{a},{b}] does not contain the root of the function.')
    
print(f"Root using Secant Method = {x_next:.4f}, Iterations = {s_iter}")
print(f"Root using Bisection Method = {c:.4f}, Iterations = {b_iter}")
```
