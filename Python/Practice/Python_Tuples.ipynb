### Python Tuples

##### Creating Tuples

```python
#Empty tuple

t1 = ()
type(t1)
```

```python
# Tuple with one element

t2 = (2)
print(type(t2))

t2 = (2,)
print(type(t2))
```

```python
t3 = tuple([1,2,3,4])

print(t3)
print(type(t3))
```

##### Accessing elements

```python
t = (1,2,3,'Hello',[4,5,6],(9,True,3+4j))

print(t[4][0])
print(t[5][2])
print(t[5][2].real,t[5][2].imag)
```

##### Operations on Tuple

```python
# Concatenation

t1 = (1,2,3)
t2 = (4,5,6)

t = t1 + t2

print(t)
print(id(t),id(t1),id(t2))
```

```python
# Repetition

t = t1*2
t
```

```python
#Loop

for i in t1:
    print(i)
```

```python
#Membership

t = (1,2,3,'Hello',[4,5,6],(9,True,3+4j))

print('Hello' in t)
print((3+4j) in t) #in operator does not check nested elements
```

##### Tuple Functions

```python
t = (1,2,3,'Hello',[4,5,6],(9,True,3+4j))

print(len(t)) #possible in heterogenous and homogenous tuples

#possible only in homogenous tuples

t1 = ('Hello','hi','Red')
t2 = (1,2,5,7,8)

print(sorted(t1))
print(max(t1))
print(min(t1))

print(sorted(t2))
print(max(t2))
print(min(t2))

#possible only if all elements of tuples are numbers

print(sum(t2))
```

```python
# count

t = (1,2,3,4,5,1)

print(t.count(50))
print(t.count(1))
```

```python
#index

print(t.index(3))
print(t.index(50)) #error
```

##### Tuple Unpacking

```python
(a,b) = (1,2)

print(a,b)

(x,*y,z) = (1,2,3,4,'hello')
print(x,y,z)

t = (1,2,3,4)
a,b,*c = t
print(a,b,c)
```

### zipping tuples

```python
a = (1,2,3,4)
b = (5,6,7,8)

tuple(zip(a,b))
```
