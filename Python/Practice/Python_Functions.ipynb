##### *args

```python
def multiply (*args):
    pro=1
    for i in args:
        pro=pro*i
    return pro

print(multiply(1,2,3,4))
```

```python
def a(*args):
    return sum(list(args))

p = a(10,20,19)
p
```

##### *kwargs

```python
def capitals(**kwargs):
    for i,j in kwargs.items():
        print(i,'->',j)

capitals(Nepal='Kathmandu',France='Paris')
```

```python
def a(**kwargs):
    return kwargs

p = a(aa=10,b=20,c=19)
p
```

##### Qn: Write a function `student_report(*marks, **details)` that accepts any number of marks and any number of student details (like Name, Age, City), then prints the details, total marks, average marks, and number of subjects.


```python
def student_report(*marks , **details):
    print('Student Detail:')
    for i,j in details.items():
        print(i,'->',j)

    print()

    total = sum(marks)
    no_of_subjects = len(marks)

    if no_of_subjects == 0:
        print('No marks provided. Returning...')
        return

    print(f'Total Marks: {total} and Average: {total/no_of_subjects}')
    print()
    print(f'No of Subjects: {no_of_subjects}')

student_report(20,50,80,90,18,Name = 'Ashuka', Gender = 'Female', Address = 'Naya Thimi')
```

##### Function as a First Class Citizen

```python
#reduce->applies function of 2 arguments cummulatively to items of a sequence,reducing it to a single value

from functools import reduce

a=[1,2,3,4,5,6,7]

print(reduce(lambda x,y:x+y,a))

print(reduce(lambda x,y:x+y,a,2))
```
