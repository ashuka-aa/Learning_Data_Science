### Python Sets

##### Creating Sets

```python
# empty sets

s ={}
print(type(s))

s1 = set()
print(s1)
print(type(s1))
```

```python
# 1D set
s1 = {1,2,3}
print(s1)

s2 = {1,2,3,{4,5}}  # set can't have mutable items
print(s2)

```

```python
# duplicates not allowed
s5 = {1,1,2,2,3,3}
print(s5)

set1={0,False}
set1
```

```python
# homogeous and hetrogeneous sets

s1 = {1,2,3}
print(s1)

s3 = {1,'hello',4.5,(1,2,3)}
print(s3)
```

```python
# using type conversion
s4 = set([1,2,3])
print(s4)
```

```python
s1 = {1,2,3}
s2 = {3,2,1}

print(s1 == s2) #sets are unordered
```

##### Adding items

```python
s = {1,2,3}
s.add(5) #can add only one item
print(s)
```

```python
#Adding multiple items at once

s.update([1,2,3])
s
```

##### Deleting items

```python
s ={1,2,3,4,5,6,True,'Hello'}

s.discard(5) 
print(s)

s.discard(50) #does not throw error if the element is not present in the set
print(s)
```

```python
s.remove(2)
print(s)

s.remove(20) # throws error if the element is not present in the set
```

```python
s.pop() #delets and returns a random item from the set
```

```python
s.clear() #empties the set
print(s)
```

```python
s = {1,2,3}

del s #deletes the whole set

print(s)
```

##### Set Operations

```python
# Union

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8,9}

print(s1 | s2)
```

```python
# Intersection
s1 & s2
```

```python
# Difference

s1 - s2
```

```python
# Symmetric Difference (union - intersection)

s1 ^ s2
```

```python
# Membership Test

1 not in s1
```

```python
# Iteration

for i in s1:
  print(i)
```

##### Set Functions

```python
len(s1)
```

```python
min(s1)
```

```python
max(s1)
```

```python
sum(s1)
```

```python
set1 = {0,5,10,True} #only possible in homogenous sets

print(sorted(set1))
print(sorted(set1, reverse = True))
```

```python
s1.union(s2) #returns a new set
```

```python
s1.update(s2) # stores the union of s1 and s2 in set s1

s1
```

```python
# intersection/intersection_update

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.intersection(s2)

s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# difference/difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.difference(s2)

s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# symmetric_difference/symmetric_difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.symmetric_difference(s2)

s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint/issubset/issuperset
s1 = {1,2,3,4}
s2 = {7,8,5,6}

s1.isdisjoint(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5}

s1.issuperset(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5}

s2.issubset(s1)
```

```python
# copy
s1 = {1,2,3}
s2 = s1.copy()

print(s1)
print(s2)
```

##### Frozenset

```python
# create frozenset
fs1 = frozenset([1,2,3]) # immutable version of a Python set object
fs1
```

###### All the read functions of set like union, intersection, difference, symmetric difference, isdisjoint, issubset, issuperset works on Frozenset
###### But write functions like update, add, intersection_update, difference_update, symmetric difference_update, delete does not work on frozenset
###### 2D sets were not possible in sets as sets can only contain immutable objects but 2D frozensts are poosible as it is immutable

```python
fs2 = frozenset([4,5,6])

fs2 | fs1
```
