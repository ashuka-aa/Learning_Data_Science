#### Creating a list

```python
# empty list

l=[]
print('Empty list:',l)
```

```python
# List with Elements

L=[1,2,3,4] #1D list
M=[1,2,3,[4,5]] #2D list
N=[[[1,2],[3,4]],[[12,14],[15,90]]] #3D list
O=[1,True,'Hii'] #Heterogeneous list

print(L,M,N,O,sep='\n')
```

```python
# using list functions

print(list('Lemon'))
print(list(range(1,5)))
```

```python
# from user input

#input as a single line 
l=input('Enter a string:').split()
print(l)

#input numbers and convert into integers
l=list(map(int,input('Enter numbers:').split()))
print(l)

#using list comprehension
l=[int(x) for x in input('Enter numbers:').split()]
print(l)

#using loops
l=[]
n=int(input('Enter the number of itmes in a list:'))

for i in range(n):
    item=input('Enter an item:')
    l.append(item)
print(l)
```

#### Accessing Items from a List

```python
#through indexing

L=[1,2,3,4]
print(L[0])   #positive indexing
print(L[-1])  #negative indexing

#through slicing

print(L[0:3])
print(L[-3:])
print(L[0::2])
```

#### Adding Items to a List

```python
l=[1,2]
l.append(5) #adds 5 at the end
print(l)
l.append([2,3]) #appends the list [2,3] as a single element
print(l)
print()

l.extend([6,7]) #adds 6 and 7 individually 
print(l)
l.extend([[8,9]]) #adds [8,9] as a single element as it is inside another list
print(l)
l.extend('hi') #adds each character 'h' and 'i' individually as string is iterable
print(l)
l.extend(['hi']) #adds 'hi' as a single element as it is inside a list
print(l)
print()

l.insert(1,80) #inserts 80 at index 1 the rest shifts to the right
print(l)
```

#### Editing Items in a List

```python
l=[1,2,3,4]

#editing with indexing 
l[2]=20 #replaces the item at index 2 with 20
print(l)

#editing with slicing
l[1:3]=[18,19] #replaces the item at index 1 and 2 with 18 and 19

print(l)
```

#### Deleting Items from a List

```python
l=[1,2,3,1,4,5,6,7,8]
print(l)

#del
del l[5] #using indexing
print (l) 
del l[1:3] #using slicing
print(l)

#remove 
l.remove(1)
print(l)

#pop 
last=l.pop(1)
print(l)
print('The popped item',last)
last=l.pop()
print(l)
print('The popped item',last)

#clear 
l.clear()
print(l)
```

#### Operations on a list

```python
#Concatenation

l=[1,2,3,4,5]
m=[1,2,3,4,5,[6,7]]
print(l)
print(m)
print(l+m)
```

```python
#Repetition

l=[1,2,3,4,5]
print(l*2)
```

```python
# Membership

l=[1,2,3,4,5]
m=[1,2,3,4,5,[6,7]]
print(5 in l)
print(6 in m)
print([6,7] in m)
```

```python
# Loops

m=[1,2,3,4,5,[6,7]]
for i in m:
    print(i)
print()
l=['apple','watermelon','orange']
while l:
    print(l.pop())
```

#### Essential List Methods in Python

```python
l=[1,9,0,7,5,4,'Hii']
print(len(l)) 

l=[1,9,0,7,5,4]
print(max(l)) 
print(min(l))
print(sorted(l))
print(sorted(l,reverse=True))

l=['hi','i','am','fine']
print(max(l)) 
print(min(l))
print(sorted(l))
print(sorted(l,reverse=True))
```

```python
l=[1,9,0,7,5,1,4,0]
print(l.count(0))

print(l.index(1))
#print(l.index(10)) throws an error as 10 is not on the list
```

```python
l=[1,9,0,7,5,4,0]
print(l)
print(id(l))

l1=l.copy() #creates a shalllow copy
print(l1)
print(id(l1))
```

```python
l=[1,9,0,7,5,4,0]
l.reverse()
print(l)

l.sort()
print(l)
l.sort(reverse=True) #in descending order
print(l)  

#reverse and sort permanently changes the list
```

#### Traversing a List

```python
#item-wise
basket=['apple','banana','guava','pineapple']
for fruit in basket: 
    print(fruit)

#index-wise
l=[1,2,3,4]
for i in range(0,len(l)):
    print(l[i])

#enumerate
fruits=['kiwi','apple','watermelon','banana']
for i,f in enumerate(fruits):
    print(f'index {i}:item {f}')
```

#### Combining Lists Using zip()

```python
l1=[1,2,3,4]
l2=[5,6,7,8,9]
print((zip(l1,l2)))
print(list(zip(l1,l2)))

l=[i+j for i,j in zip(l1,l2)] #adding two lists element-wise
print(l)
```

```python
l=[1,2,3,4]
cube=[i**3 for i in l]
print(cube)
```

##### Qn: WAP to enter elements in a list and print two separate list containing even and odd elements.

```python
list1 = [None] * 5

for i in range(len(list1)):
    a = int(input(f'Enter the {i+1} element:'))
    list1[i] = a

list_even = []
list_odd = []

for j in list1:
    if j % 2 == 0:
        list_even.append(j)
    else:
        list_odd.append(j)

print(f'Even list: {list_even}')
print(f'Odd list: {list_odd}')
```

##### Qn: Given a list of words, create a list of their first letters

##### Qn: Filter out even numbers form a given list

```python
numbers=[2,5,6,9,0,11,15]
even=[x for x in numbers if x%2==0]

print(even)
```

##### Qn: Label the numbers of a list as Even or Odd

```python
numbers=[2,5,6,9,0,11,15]
even_or_odd=['Even' if x%2==0 else 'Odd' for x in numbers ]

print(even_or_odd)
```

##### Qn: Flatten a 2D list

```python
L=[[1,2],[3,4],[5,6]]
L1=[x for row in L for x in row]

print(L1)
```

##### Qn: Creating a 3D list

```python
matrix=[[[i*j*k for i in range(1,3)] for j in range(1,3)] for k in range(1,3)]
print(matrix)
```
