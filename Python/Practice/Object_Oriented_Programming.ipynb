#### Qn: Create your own data type Fraction in Python that represents rational numbers and supports arithmetic operations.

```python
class Fraction:

    def __init__(self,x,y):
        self.num=x
        self.den=y
        self.simplify()

    def __str__(self):
        if self.den==1:
            return f'{self.num}'
        else:
            return f'{self.num}/{self.den}'
    
    def __add__(self,other):
        new_num= self.num * other.den + self.den * other.num
        new_den=self.den * other.den
        return Fraction(new_num,new_den)
        
    def __sub__(self,other):
        new_num= self.num * other.den - self.den*other.num
        new_den=self.den * other.den
        return Fraction(new_num,new_den)
        
    def __mul__(self,other):
        new_num= self.num * other.num
        new_den=self.den * other.den
        return Fraction(new_num,new_den)

    def __truediv__(self,other):
        new_num= self.num * other.den 
        new_den=self.den * other.num
        return Fraction(new_num,new_den)
    
    def simplify(self):
        import math
        a=math.gcd(self.num,self.den)
        self.num//=a
        self.den//=a

obj=Fraction(12,4)
obj2=Fraction(3,10)

print(obj)
print(obj2)
```

#### Qn: Calculate the factorial of a number

```python
class Fact:
    def facto(self,n):
        if n == 0 or n == 1:
            return 1

        else:
            return n * self.facto (n-1)

def main():
    num = int(input('Enter a number:'))

    if num < 0:
        print('Factorial does not exist')

    else:
        f = Fact()
        r = f.facto(num)

        print(f'{num}! = {r}')

if __name__ == '__main__':
    main()
```

#### Qn: Calculate the fibonacci series upto nth term

```python
class Fibo:
    def fibo(self,n):
        if n == 1 :
            return 1
        elif n == 0:
            return 0
        else:
            return (self.fibo(n-2) + self.fibo(n-1))

def main():
    num=int(input('Enter the number:'))
    
    if num<=0:
        print('Cannot be computed')
    
    else:
        f1 = Fibo()
        
        for i in range(1,num+1):
            print(f1.fibo(i),end=' ')

if __name__ == '__main__':
    main()
```

#### Using *args and **kwargs

```python
class Profile:
    def __init__(self,**kwargs):
        self.data= kwargs
        for key, value in kwargs.items():
            setattr(self,key,value)

    def display (self):
        for key,value in self.data.items():
            print(f'{key}:{value}')

p1 = Profile(name='Ashuka',age=20,city='Bkt')
p1.display()
print()
p2 = Profile(name='Yoyo',age=2,city='Ktm')
p2.display()
```

```python
class Mixed:
    def __init__(self,*args,**kwargs):
        self.all = list(args)
        self.d = kwargs

    def display(self):
        print(f' List: {self.all}, Dict: {self.d}')

m1 = Mixed(10,20,'hello', True, name = 'Ashuka',age= 20)
m1.display()
```

```python
class Person:
    def __init__(self,name):
        self.name= name

    def details(self,**kwargs):
        for key,value in kwargs.items():
            setattr(self,key,value)
            # print(key,value)

    def display(self):
        for key,values in self.__dict__.items():
            print(f'{key}:{values}')

p1 = Person('Ashuka')

p1.details(Age = 20, Address = 'Bkt')
p1.details(Age = 21,Email= 'Ashuka@gamil.com')
p1.display()
```

#### Qn: Design a class Atm in Python that allows a user to set a PIN and balance, and perform operations like changing PIN, checking balance, and withdrawing money using a menu-driven system.

```python
class Atm:
    
    def __init__(self):
        self.pin=0
        self.balance=0
        self.set()
        
    def set(self):
        pin=int(input('Enter your pin: '))
        balance=int(input('Enter your balance: '))
        self.pin=pin
        self.balance=balance
        self.menu()

    def menu(self):
        user_input=input(''' Hi what do you want to do?
        1)Press 1 to Change pin
        2)Press 2 to Check balance
        3)Press 3 to Withdraw
        4)Press anything else to Exit
        ''')

        if user_input == '1':
            self.change_pin()
        elif user_input == '2':
            self.check_balance()
        elif user_input == '3':
            self.withdraw()
        else:
            print('Exiting ATM......')
            return

    def change_pin(self):
        old=int(input('Enter your old pin: '))

        if old==self.pin:
            new=int(input('Enter your new pin: '))
            self.pin=new
            print('Pin changed successfully.')
        else:
            print('Your pin is incorrect.')
        self.menu()

    def check_balance(self):
        input_pin=int(input('Enter your pin: '))
        if input_pin==self.pin:
            print('Your balance is:',self.balance)
        else:
            print('Your pin is incorrect.')
        self.menu()

    def withdraw(self):
        input_pin=int(input('Enter your pin: '))
        if input_pin==self.pin:
            amount=int(input('Enter the amount you want to withdraw: '))

            if amount<=self.balance:
                self.balance-=amount
                print('Amount withdrawn successfully.')
                print('Your new balance is:',self.balance)
            else:
                print('Insufficient balance.')
        else:
            print('Your pin is incorrect.')
        self.menu()


a=Atm()
```

```python
#using super keyword

class Phone():
    def __init__(self,price,brand,model):
        print('Inside phone constructor')
        self.price=price
        self.brand=brand
        self.model=model
    
class SmartPhone(Phone):
    def __init__(self, price,brand,model,os, ram):
        print('Inside smartphone constructor')
        super().__init__(price,brand,model)
        self.os=os
        self.ram=ram
        print('Inside smartphone constructor')
    
s=SmartPhone(2000,'Iphone',17,'Windows','8GB')
```

* Multi-level Inheritance

```python
class Grandparent:
    def house(self):
        print('Grandfather owns a house')
class Parent(Grandparent):
    def car(self):
        print('Parents owns a car')
class Child(Parent):
    def bike(self):
        print('Child owns a bike')

c = Child()
c.house()
c.car()
c.bike()
```

* Mutiple Inheritance

```python
class A:
    def show(self):
        print('Inside A')
        super().show()
        

class B:
    def show(self):
        print('Inside B')
        

class C(A,B):
    def show(self):
        print('Inside C')
        super().show()

c1 = C()

c1.show()

print(C.mro())
```

* Hierarchical Inheritance

```python
class A:
    def show(self):
        print('Inside Class A')

class B(A):
    def show(self):
        print('Inside Class B')
        super().show()

class C(A):
    def show(self):
        print('Inside Class C')
        super().show()

c = C()
c.show()

print(C.mro())

b = B()
b.show()

print(B.mro())
```

* Hybrid Inheritance

```python
class M1:
    def display(self):
        print('Class M1')
        super().display()

class A(M1):
    def display(self):
        print('Class A')
        super().display()

class M2:
    def display(self):
        print('Class M2')

class B(M2):
    def display(self):
        print('Class B')
        super().display()

class C(A,B):
    def display(self):
        print('Class C')
        super().display()

d = C()

d.display()

print(C.mro())
        
```

```python
#Diamond Problem

class A():
    def  show(self):
        print('Class A')

class B(A):
    def show(self):
        print('Class B')
        super().show()

class C(A):
    def show(self):
        print('Class C')  
        super().show()

class D(B,C): #whichever parent class is written first that is next in the MRO
    def show(self):
        print('Class D')
        super().show()
a=A()
d=D()
d.show()

print(D.__mro__)
```
