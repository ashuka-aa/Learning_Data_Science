```python
#WAP to find the factorial of a given number using for loops

num = int(input('Enter a number'))
fact = 1
if num < 0:
    print('Factorials are only defined for +ve integers ')
else:
    for i in range(1,num+1):
        fact *= i

    print(f'{num}! = {fact}')
```

```python
#WAP to sum of natural numbers using for loop

num = int(input('Enter a number'))
result = 0

if num<=0:
    print(f'{num} is not a natural number')
else:
    for i in range(1,num+1):
        result+=i
    
    print(f'Sum of first {num} natural numbers = {result}')
```

```python
# Reverse of a number

num = 1234
rev = 0
while num > 0:
     r = num % 10
     rev = rev * 10 + r
     num//=10

print(f'The reverse is {rev}')
```

```python
# Simple Guessing Game using while else statement

import random
correct=random.randint(1,100)
count=1
guess=int (input('Guess a number: '))
while guess!= correct:
    if (guess<correct):
        print('Guess higher:')
    else:
        print('Guess lower: ')

    guess=int(input('Next guess: '))
    count+=1

else:
    print(guess,' is the correct guess.')
    print('You guessed it in',count,'times.')
```

```python
# Nested Loops

r=int(input('Enter the number for generating a pyramid: '))

for i in range(0,r+1):
    print(' '* 2*(r-i), end='')
    for j in range(i,-1,-1):
        print(j, end=' ')
    for k in range(1,i+1):
        print(k, end=' ')
    print()
```

```python
#break statement

l=10
u=30

for i in range(l,u+1):
    for j in range(2,i):
        if i % j==0:
            break
    else:
            print(i)
```
