## Python Dictionary

#### Creating a Dictionary

```python
# empty dictionary
d = {}
d
```

```python
# 1D dictionary
d1 = {
    'Name':'Ashuka',
    'Roll No':10,
    'Address': 'Naya Thimi',
}
print(type(d1))
print(d1)
```

```python
# with mixed keys
d2 = {(1,2,3):1,'hello':'world'}
d2
```

```python
# 2D dictionary

d3 = {
    'Name': 'Ashuka',
    'Gender':'Female',
    'Marks':{
        'DSA':92,
        'AI':80,
        'Statistics':90
    }
}
d3
```

```python
# using dict function
d4 = dict((('Name','Ashuka'),('Gender','Female')))
d4
```

##### Dictionary Comprehension

```python
# print 1st 10 numbers and their squares

d={i:i**2 for i in range(1,11)}
print(d)
```

```python
d={'one':1,'two':2,'three':3}
d1={i:j**2 for i,j in d.items()}
print(d1)
```

```python
l1=['one','two','three','four','five']
l2=[1,2,3,4,5]

d2={i:j for i,j in zip(l1,l2)}
print(d2)
```

```python
stock={'TV':10,'Laptop':0,'Fridge':20}
d1={key:value for  (key,value) in stock.items() if value > 0}
print(d1)
```

```python
# print tables of number from 2 to 4

d={key:{i:i*key for i in range(1,11)} for key in range (2,5)}
print(d)
```
