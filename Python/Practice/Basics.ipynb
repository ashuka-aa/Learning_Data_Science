### Python Operators

```python
# Arithmetic Operators

x = int(input('Enter the value of x:'))
y= int(input('Enter the value of y:'))

print(x+y, x-y, x*y, x%y, x/y, x//y, x**y,sep='\n')
```

```python
# Relational Operators

a = 10
b = 15

print(a < b)
print(a > b)
print(a == b)
print(a <= b)
print(a >= b)
print(a != b)
```

```python
# Logical Operators

# '' is considered false and if the string contains any character then it is considered true

# 'and' returns the last true value if both are true and first false value if there is a false value
print('Using "and" operator')
print(2 and 3)
print('' and 'hello') 
print('hello' and 'hi')
print('hi' and '')
print('' and '')

# 'or' returns the first true value and last false value if there is a false value
print('Using "or" operator')
print(2 or 3)
print('' or 'hello') 
print('hello' or 'hi')
print('hi' or '')
print('' or '')

# not returns false if true and vice-versa
print('Using "not" operator')
print(not '')
print(not 3)
print(not 'hello')
```

```python
# Bitwise Operators

#They change the inputs in binary, perform the operations and then convert it back to decimal

#Bitwise and
print(2 & 3)

#Bitwise or 
print(2 | 3)

#Bitwise not
print(~3)

#Bitwise XOR
print(2 ^ 3)

#Bitwise left shift 
print(2 << 3)

#Bitwise right shift 
print(15 >> 3)
```

```python
# Unary Operators

x = 5
print(-x) #unary minus
print(+x) #unary plus
```

```python
#Assignment Operator

a = 1
b = 3
print(a,b)
```

```python
#Membership Operator

print('D' in 'Dance')
print('d' in 'Dance')

print('f' not in 'Aeroplane')
print('e' not in 'Aeroplane')
```

```python
#is operator

a = [1,2,3,10]
b = a

print(b is a)

c = [1,2,3,10]

print(c is a)

print(id(a),id(b),id(c)) #a and b points to the same object and c is a diff. object
```

### Python Conditionals

```python
#Checking whether the number is even or odd
x = int(input('Enter a number'))

if x>0:
    if x%2==0:
        print(f'{x} is even')
    else:
        print(f'{x} is odd')
else:
    print(f'{x} is negative')
```

```python
#find profit or loss

cp=float(input('enter the cost price '))
sp=float(input('enter the selling price '))

p=sp-cp

if p>0:
    print('Profit =',p)
elif p==0:
    print('No profit, No loss')
else:
    print('Loss =',p)
```

```python
# WAP making an if-else ladder

marks=abs(float(input('Enter your marks:')))

if marks >= 80 and marks <= 100:
    print('Congratulations')
elif marks < 80 and marks >= 70:
    print('First division')
elif marks < 70 and marks >= 60:
    print('Second division')
elif marks < 60 and marks >= 50:
    print('Third division')
else:
    print('Failed')
```

##### WAP to make if-else ladder, if no > 90 then multiply with its reciprocal, if no >= 50 then divide it by 20, if no > 20 then multiply it by pi and find the sinosudal and cosine function of that value, else print number is negative and we can't do anything 

```python
num = int(input('Enter a number:'))

if num > 90:
    r = num * (1/num)
    print('-'*60)
    print(f'After multiplying {num} with its reciprocal: {r}')
    print('-'*60)
elif num >= 50:
    r = num/20
    print('-'*60)
    print(f'After dividing {num} by 20: {r}')
    print('-'*60)
elif num > 20:
    import math
    r = math.pi * num
    sin_val = math.sin(r)
    cos_val = math.cos(r)
    print('-'*60)
    print(f'The sin value = {sin_val} and cos value = {cos_val}')
    print('-'*60)
else:
    print(f'{num} is negative or less than or equal to 20')
```
